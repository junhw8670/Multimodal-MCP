#**1. 지도학습과 강화학습**
###**1.  지도학습 (Supervised Learning)**
지도학습은 정답(라벨)이 포함된 데이터로 모델을 학습시키는 방식입니다. 즉, 입력 XXX와 그에 대응하는 정답 YYY를 함께 제공하여, 모델이 X→YX \rightarrow YX→Y로 매핑하는 함수를 학습하도록 합니다. 학습 과정에서는 모델의 예측값과 실제 정답 사이의 오차(loss)를 계산하고, 이를 최소화하도록 가중치를 업데이트합니다. 대표적인 문제 유형은 분류(Classification)와 회귀(Regression)이며, 이미지 분류, 스팸 메일 판별, 주가 예측 등이 이에 해당합니다. 핵심은 “정답이 있는 데이터로 오차를 줄이며 학습한다”는 점입니다.

###**2. 강화학습 (Reinforcement Learning)**
강화학습은 정답을 직접 주지 않고, 행동에 대한 보상(Reward)을 통해 학습하는 방식입니다. 에이전트(agent)가 환경(environment)과 상호작용하며 행동(action)을 선택하고, 그 결과로 받은 보상을 최대화하는 방향으로 정책(policy)을 학습합니다. 즉, “이 행동이 좋았는가?”를 보상으로 평가받으며 점점 더 좋은 전략을 찾아갑니다. 게임 AI, 자율주행, 로봇 제어 등이 대표적인 예입니다. 핵심은 “정답 대신 보상을 통해 최적 행동 전략을 학습한다”는 점입니다.

#**2. 순차적 의사결정**
에이전트가 한 번의 선택이 아니라 시간의 흐름에 따라 여러 번 행동을 선택하고, 그 결과가 이후 상태와 보상에 영향을 미치는 구조를 의미합니다. 즉, 현재의 행동이 단순히 즉각적인 보상만 결정하는 것이 아니라, 미래의 상태 전이와 장기적인 누적 보상에까지 영향을 주기 때문에 에이전트는 매 순간의 보상보다 장기적 보상(Return)을 최대화하도록 정책을 학습합니다. 이러한 구조는 보통 상태(State), 행동(Action), 보상(Reward), 상태전이(Transition)로 구성된 마르코프 결정 과정(MDP)으로 모델링되며, 과거 전체가 아니라 현재 상태에 기반해 의사결정을 한다는 마르코프 성질을 가정합니다.

#**3. 리워드**
리워드(Reward)란, 에이전트가 특정 상태에서 어떤 행동을 했을 때 환경으로부터 받는 즉각적인 피드백 신호입니다. 이는 행동의 좋고 나쁨을 수치로 표현한 값으로, 에이전트는 이 리워드를 기반으로 자신의 정책(Policy)을 업데이트하며 장기적인 누적 보상(Return)을 최대화하도록 학습합니다. 리워드는 매 타임스텝마다 주어질 수도 있고, 에피소드 종료 시 한 번만 주어질 수도 있으며, 양수는 바람직한 행동을, 음수는 바람직하지 않은 행동을 의미하는 경우가 일반적입니다. 중요한 점은 강화학습의 목표가 단순히 현재 리워드를 최대화하는 것이 아니라, 할인율(γ)을 고려한 미래 보상까지 포함한 누적 리워드를 최대화하는 것이라는 점입니다.

###**1. 즉각적 피드백(Immediate Signal)**
리워드는 특정 상태에서 행동을 했을 때 즉시 주어지는 수치형 피드백입니다. 이 값은 해당 행동이 얼마나 바람직했는지를 나타냅니다. 예를 들어, 게임에서 점수를 얻으면 +보상, 장애물에 부딪히면 -보상을 받는 식입니다. 다만, 강화학습의 목표는 이 즉각적 보상 하나가 아니라 장기적 누적 보상(Return)을 최대화하는 것입니다.



###**2. 희소성(Sparse Reward)**
리워드는 매 순간 주어지지 않을 수도 있습니다. 많은 환경에서는 대부분의 시간 단계에서 보상이 0이고, 특정 목표를 달성했을 때만 보상이 주어집니다. 이를 희소한 리워드라고 합니다. 예를 들어, 미로 문제에서 출구에 도착했을 때만 +1을 받는 경우가 이에 해당합니다. 희소할수록 학습이 어려워집니다.



###**3. 지연성(Delayed Reward)**
어떤 행동의 결과에 대한 보상이 즉시 나타나지 않고, 여러 단계를 거친 뒤 나중에 주어질 수 있습니다. 이를 지연된 리워드라고 합니다. 예를 들어, 체스에서는 한 수를 둔 직후 보상이 없고, 게임이 끝난 뒤 승패로 보상이 주어집니다. 이로 인해 어떤 행동이 최종 결과에 기여했는지를 파악하기 어려워지며, 이를 신용 할당 문제(Credit Assignment Problem)라고 합니다.

#**4. 에이전트**
에이전트(Agent)란, 환경(Environment)과 상호작용하며 상태(State)를 관찰하고, 그에 따라 행동(Action)을 선택하고, 그 결과로 리워드(Reward)를 받아 학습하는 주체를 의미합니다. 에이전트의 목표는 단기적인 보상이 아니라 미래 보상을 포함한 누적 보상(Return)을 최대화하는 정책(Policy)을 학습하는 것입니다. 이를 위해 에이전트는 경험을 통해 가치 함수(Value Function)나 정책을 점진적으로 업데이트하며, 탐험(Exploration)과 활용(Exploitation)을 균형 있게 수행합니다.

###**1. 가치 기반 에이전트 (Value-Based Agent)**
가치 함수(Value Function)를 학습하여 행동을 선택하는 방식입니다. 즉, 각 상태 또는 상태-행동 쌍의 가치를 계산하고, 가치가 가장 높은 행동을 선택합니다. 대표적인 알고리즘으로는 Q-Learning, DQN 등이 있습니다. 이 방식은 정책을 직접 학습하지 않고, 가치 함수를 통해 간접적으로 최적 정책을 도출합니다.



###**2. 정책 기반 에이전트 (Policy-Based Agent)**
행동 선택 규칙인 정책(Policy)을 직접 학습하는 방식입니다. 확률적으로 행동을 출력하는 정책을 최적화하며, 대표적으로 REINFORCE, Policy Gradient 방법이 있습니다. 연속적인 행동 공간(예: 로봇 제어)에 적합하다는 장점이 있습니다.



###**3. 액터-크리틱 에이전트 (Actor-Critic Agent)**
가치 기반과 정책 기반을 결합한 방식입니다.

- Actor는 정책을 학습하여 행동을 선택하고
- Critic은 가치 함수를 학습하여 Actor의 행동을 평가합니다.
- 대표적으로 A2C, A3C, PPO, SAC 등이 있습니다. 학습 안정성과 효율성이 높아 실무에서 많이 사용됩니다.


###**4. 모델 기반 에이전트 (Model-Based Agent)**
환경의 동작 원리(상태 전이 확률, 보상 함수)를 학습하거나 알고 있다고 가정하고, 이를 기반으로 미래를 예측하며 계획(Planning)을 수행하는 방식입니다. 예를 들어, MCTS(Monte Carlo Tree Search)나 Dyna-Q 등이 있습니다. 샘플 효율성이 높은 장점이 있습니다.



###**5. 모델 프리 에이전트 (Model-Free Agent)**
환경의 내부 구조를 모른 채, 경험을 통해 직접 정책이나 가치 함수를 학습하는 방식입니다. 대부분의 현대 강화학습 알고리즘(DQN, PPO 등)이 이에 해당합니다. 구현이 비교적 단순하지만, 많은 데이터가 필요합니다.

#**5. Exploration**
Exploration(탐험)이란, 에이전트가 현재까지 학습한 최적 행동만 반복하는 것이 아니라, 더 나은 보상을 얻을 가능성이 있는 새로운 행동을 시도해보는 과정을 의미합니다. 이는 장기적인 누적 보상을 최대화하기 위해 필수적인 전략으로, 이미 잘 알려진 행동을 선택하는 Exploitation(활용)과 균형을 이루어야 합니다. 예를 들어, ε-greedy 전략에서는 일정 확률(ε)로 무작위 행동을 선택하여 새로운 가능성을 탐색합니다. 탐험이 부족하면 지역 최적해에 머물 수 있고, 반대로 지나치면 학습이 불안정해질 수 있기 때문에 적절한 균형이 중요합니다.

#**6. Exploitation**
Exploitation(활용)이란, 에이전트가 지금까지의 학습 결과를 바탕으로 현재 가장 높은 보상을 줄 것으로 예상되는 행동을 선택하는 과정을 의미합니다. 즉, 이미 경험을 통해 가치가 높다고 판단된 행동을 반복하여 안정적으로 보상을 얻는 전략입니다. 이는 단기적으로는 높은 성과를 보장하지만, 새로운 가능성을 탐색하지 않기 때문에 더 나은 전략을 발견하지 못하고 지역 최적해에 머무를 위험이 있습니다. 따라서 강화학습에서는 Exploration(탐험)과 Exploitation(활용)의 균형을 맞추는 것이 핵심 과제입니다.

###**💬  일상 속에서의 Exploration VS Exploitation**
일상 속에서도 Exploration(탐험)과 Exploitation(활용)은 자연스럽게 나타납니다. 예를 들어, 점심 메뉴를 고를 때를 생각해볼 수 있습니다. 늘 가던 맛집에서 먹는 것은 Exploitation입니다. 이미 맛있다는 것을 알고 있으므로 안정적으로 만족을 얻을 수 있습니다. 반면, 한 번도 가보지 않은 새로운 식당에 도전하는 것은 Exploration입니다. 혹시 더 맛있는 곳을 발견할 수도 있지만, 실패할 위험도 있습니다.

또 다른 예로는 공부 방법을 들 수 있습니다. 지금까지 성적이 잘 나왔던 공부 방식을 계속 사용하는 것은 Exploitation이고, 새로운 학습 방법(예: 스터디 참여, 다른 교재 사용, AI 도구 활용 등)을 시도해보는 것은 Exploration입니다. 새로운 방법이 더 효율적일 수도 있지만, 그렇지 않을 수도 있습니다.
